# Data Cleaning

```python
df.drop_duplicates(inplace=True) #Remove duplicate rows 
```

```python
df = df[(df['pickup_latitude'] != df['dropoff_latitude']) | (df['pickup_longitude'] != df['dropoff_longitude'])]
#Remove rows with same pickup and dropoff location
```

```python
df = df[df['passenger_count'].fillna(0).astype(int) > 0] #Remove trips with missing or zero passenger count
```

```python
df.shape #After data cleaning
```

```python
df.describe()
```

# Data Transformation

```python
df['pickup_datetime'] = pd.to_datetime(df['pickup_datetime'])
df['pickup_month'] = df['pickup_datetime'].dt.month
df['pickup_day_name'] = df['pickup_datetime'].dt.day_name()
df['pickup_hour'] = df['pickup_datetime'].dt.hour
```

```python
df['dropoff_datetime'] = pd.to_datetime(df['dropoff_datetime'])
df['dropoff_month'] = df['dropoff_datetime'].dt.month
df['dropoff_day_name'] = df['dropoff_datetime'].dt.day_name()
df['dropoff_hour'] = df['dropoff_datetime'].dt.hour
```

```python
df["pickup_hour"] = df["pickup_hour"].astype("int64")
df["dropoff_hour"] = df["dropoff_hour"].astype("int64")
df["trip_duration"] = df["trip_duration"].astype("int64")
```

```python
df['passenger_count'] = df['passenger_count'].astype('category')
df['pickup_month'] = df['pickup_month'].astype('category')
df['dropoff_month'] = df['dropoff_month'].astype('category')
```

```python
df['store_and_fwd_flag'] = df['store_and_fwd_flag'].astype('category')
df['vendor_id'] = df['vendor_id'].astype('category')
```

# Insight Generation

```python
#Statistical Insights (Plotting matrix as heatmap)
numeric_df = df.select_dtypes(include=['number'])
corr_mat = numeric_df.corr()
plt.figure(figsize=(12, 10))
sns.heatmap(corr_mat, annot=True, cmap="coolwarm")
plt.title("Correlation Matrix of Numeric Variables")
plt.show()
```

```python
cont_var = 'trip_duration'
cat_var = 'passenger_count'

plt.figure(figsize=(14, 10))

plt.subplot(2, 2, 1)
sns.boxplot(x=cat_var, y=cont_var, data=df)
plt.title(f'Box Plot of {cont_var} vs {cat_var}')

plt.subplot(2, 2, 2)
sns.violinplot(x=cat_var, y=cont_var, data=df)
plt.title(f'Violin Plot of {cont_var} vs {cat_var}')

plt.subplot(2, 2, 3)
df.groupby(cat_var, observed=True)[cont_var].mean().plot(kind='bar')
plt.ylabel(f'Mean {cont_var}')
plt.title(f'Bar Chart of Mean {cont_var} by {cat_var}')

plt.subplot(2, 2, 4)
sns.stripplot(x=cat_var, y=cont_var, data=df, jitter=True)
plt.title(f'Strip Plot of {cont_var} vs {cat_var}')

plt.tight_layout()
plt.show()
```

```python
average_duration = df.groupby(['vendor_id', 'passenger_count'], observed=True).mean(numeric_only=True).reset_index()
average_duration[['vendor_id', 'passenger_count', 'trip_duration']].round(2)
```

```python
sns.barplot(x='passenger_count', y='trip_duration', hue='vendor_id', data=average_duration)
plt.title('Average Trip Duration by Passenger Count and Vendor ID')
plt.xlabel('Passenger Count')
plt.ylabel('Average Trip Duration (seconds)')
plt.show()
```
